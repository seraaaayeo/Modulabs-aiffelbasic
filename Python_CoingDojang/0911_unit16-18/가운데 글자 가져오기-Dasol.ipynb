```python
def solution(s):
    answer = ''
    
    tmp = len(s) // 2
    if(len(s)%2 == 0): #짝수
        answer = s[tmp-1:tmp+1]

    elif(len(s)%2 == 1): #홀수
        answer = s[tmp:tmp+1]
    
    return answer
```

```python
def main():
    s = "abcde"
    print(solution(s))
    s = "qwer"
    print(solution(s))
```

```python
if __name__ == "__main__":
    main()
```
